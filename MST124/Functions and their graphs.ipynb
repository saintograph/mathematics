### Functions and their graphs

Sketch the graph of the function:

$$f(x)=\frac{1}{4}x^2-2x+6 \ \ \ \ (5\leq x \lt 7)$$

Completing the squares gives:
$$ f(x) = \frac{1}{4}x^2 - 2x + 6$$
$$=\frac{1}{4}(x^2 - 8x)+6$$
$$=\frac{1}{4}((x-4)^2 -16) + 6$$
$$=\frac{1}{4}(x-4)^2 - 4 +6$$
$$=\frac{1}{4}(x-4)^2 + 2$$

```python
from sympy.plotting import plot 
from sympy import *
from fractions import Fraction
```

```python
x=Symbol('x')
y=Symbol('y') 
plot(Fraction(1, 4) * (x - 4)**2 + 2, (x, -100, 100))
```

#### Sketching exercise

1. $f(x) = 3-2x$, $(-1 < x < 4)$

The function is a linear equation, with an expected type - straight line.

When $y = 0$:

$$3-2x=0$$
$$-2x = -3$$
$$x=\frac{3}{2}$$

$y$-intercept:
$$f(x) = 3-2(0)$$
$$y=3$$

The domain is $(-1 < x < 4)$


for $x=-1$

$$f(x)=3-2(-1)$$
$$=3+2$$
$$=5$$

for $x = 4$

$$f(x)=3-2(4)$$
$$=3-8$$
$$=-5$$


```python
# Let's check!
plot((3 - 2 * x), ylim = [-6, 6], xlim = [-1, 4])
```

2. $f(x) = - \frac{1}{2} x^2 - 2x -5$, for $(x \geq -5)$

Function is quadratic, with a $n$ shaped plot because the coefficient for $x^2$ is negative.

Completed square form.

$$=-\frac{1}{2}(x^2+4x)-5$$

$$=-\frac{1}{2}((x+2)+4)-5$$

$$=-\frac{1}{2}(x+2)^2 + 2 -5$$

$$=-\frac{1}{2}(x+2)^2-3$$

Calculating the domain, where $x \geq -5$, 

$$=-\frac{1}{2}(-5+2)^2-3$$

$$=-\frac{1}{2}(9) - 3$$

$$=-\frac{9}{2} - 3$$

$$=-\frac{9}{2}-\frac{6}{2}$$

$$=-\frac{15}{2}$$

```python
# Let's check!
plot(((-0.5 * (x + 2) ** 2) - 3), ylim = [-14, 2], xlim = [-5, 10])
```

#### Question 1

Convert expression to completed square form.

$$f(x)= -(x^2-10x) - 24$$

$$= -((x-5)-25)^2 -24$$

$$=-(x-5)^2 + 1$$

The parabola is n-shaped. The vertex (highest point) is when $-(x+5)^2$ is equal to $0$, so the highest value for $=-(x-5)^2 + 1$ is $1$. This occurs when $-(x + 5) = 0$, that is, when $x=5$. 

$$-(5 - 5 ) =0$$

Also

$$f(3) =-(3 - 5)^2 + 1$$

$$=-(-2)^2 + 1$$

$$= -4 + 1$$

$$= -3$$

which gives us 
$$(3, -3)$$

And

$$f(6) = -(6-5)^2+1$$

$$=-(-1)^2 + 1$$

$$=0$$

which gives us

$$(6,0)$$

```python
plot(-x ** 2 + 10 * x -24, ylim = [-10, 1], xlim = [-10, 10])
```
